```python
# Import necesary modules 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import qutip as Q
```

# 1. Mathematical Model

The idea of this notebook is to practice our understanding of quantum composite systems by analyzing the 1D chain of spins. We start by defining the Pauli matrices for a given one-spin system:

$$\sigma^x = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} $$

$$\sigma^y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} $$

$$\sigma^z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} $$


Then we define the following operator of the composite system:

$$\sigma^{a}_j= I^{\otimes j-1}\otimes \sigma^{a} \otimes I^{\otimes N-j}$$

Where $I$ is the identity operator, $j\in\mathbb{Z} \wedge 1\geq j \leq N$ and $a\in\{x,y,z\}$. Now, given $J_x,J_y,J_z \in\mathbb{R}$, the Hamiltonian of the system is defined as:

$$H = -\frac{1}{2}\sum^{N}_{j=1} \left(J_x\sigma_{j}^x\sigma_{j+1}^x + J_y\sigma_{j}^y\sigma_{j+1}^y + J_z\sigma_{j}^z\sigma_{j+1}^z + h\sigma_{j}^z\right) \ ,$$
where $h$ is the intensity of an external magnetic field in the $z$-direction and periodic boundary conditions (PBC) are assumed.

The task is to numerically find the energy spectrum (finding the energy eigenvalues) and the eigenstates of the system known as *Heisenberg's Model*, where $J_x=J_y=J_z=J$, for different values of $J$ and $h$.

# 2. Proposing a solution for the problem

We are asked to find values of energy $\epsilon$ and states $|\epsilon\rangle$ that verify

$$H|\epsilon\rangle = \epsilon|\epsilon\rangle$$

By definition, the state $|\epsilon\rangle$ is defined by:

$$|\epsilon\rangle = \bigotimes_{j=1}^N |\epsilon_j\rangle \ ,$$

and
 
$$H = \sum_{\epsilon} \epsilon|\epsilon\rangle\langle\epsilon|$$

To solve the problem at hand, we're going to focus on getting $H$ to be represented by a 2-D matrix, so that we can apply useful eigenvalue solvers like `numpy.linalg.eig`. 

Let's begin by defining a basis of the composite Hilbert space of states $\mathbb{H}$ with dimension $\dim(H) = M^N$:

$$\{|\psi_1\rangle\otimes|\phi_1\rangle\otimes\cdots\otimes|\alpha_1\rangle, |\psi_1\rangle\otimes|\phi_1\rangle\otimes\cdots\otimes|\alpha_2\rangle,..., |\psi_M\rangle\otimes|\phi_M\rangle\otimes\cdots\otimes|\alpha_M\rangle\}$$

Any given state of $\mathbb{H}$ can be written as 

$$|\psi\rangle = c_{1,1,...,1}|\psi_1\phi_1\cdots\alpha_1\rangle + \cdots + c_{M,M,...,M}|\psi_M\phi_M\cdots\alpha_M\rangle\ ,$$

but this way of refering to the states generates the following way of refering to a given operator $H$ of $\mathbb{H}$:

$$ H = \sum_{i,j,\cdots,k \\ i',j',\cdots,k'} |\psi_i\phi_j\cdots\alpha_k\rangle H_{i,j,\cdots,k}^{i',j',\cdots,k'}\langle\psi_{i'}\phi_{j'}\cdots\alpha_{k'}| \ \ ; H_{i,j,\cdots,k}^{i',j',\cdots,k'}\equiv \langle\psi_i\phi_j\cdots\alpha_k|H|\psi_{i'}\phi_{j'}\cdots\alpha_{k'}\rangle $$

Notice this get's really hard to follow when several subsystems are involved.

On the other hand, we can represent $|\psi\rangle$ as

$$|\psi\rangle \doteq \begin{bmatrix} c_0 \\ c_1 \\ c_2 \\ \vdots \\ c_{M^N - 1} \end{bmatrix} \ ,$$

where each coefficient $c_i$ is associated with a given base state $|c_i\rangle$, for example $|c_2\rangle = |\psi_1\phi_1\cdots\alpha_2\rangle$. This mapping is particularly easy to perform when every subsystem has the same dimensionality $M$, which is the case for this problem.

For the Heisenberg model, we know that every subsystem consisting of a single spin has a dimension $M=2$. We can create a mapping for any given index $n$ to the indexes $\psi_n, \phi_n, \cdots, \alpha_n$:

$$\alpha_n = n_0\ //\  M^{N - 1}$$
$$\vdots$$
$$\phi_n = n_{N-2}\ //\ M^{1}$$
$$\psi_n = n_{N-1}\ //\ M^{0}$$

where the variable $n_i$ is recursively defined as $n_{i+1} = n_{i} - (n_{i}\ //\ M^{N-i-1})\cdot M^{N-i-1}$, and the operator $//$ represents integer division, defined as 

$$\forall a,b \in \mathbb{Z}, a\ //\ b : a = (a\ //\ b)\cdot b + a\mod b$$

Having defined the mapping, it is easy to now define and calculate the representation of $H$ as:

$$H = \sum_{i,j} |c_i\rangle H_{i,j} \langle c_j| \ \ ; H_{i,j} \equiv \langle c_i|H|c\rangle$$

And calculating the eigenenergies and eigenstates is straightforward from there.

# 3. Building the code for the simulations

```python
# 1 - DEFINE PAULI MATRICES
sigma_0 = Q.qeye(2)
sigma_x = Q.sigmax()
sigma_y = Q.sigmay()
sigma_z = Q.sigmaz()


# 2 - SOME AUXILIARY FUNCTIONS
def sigma_j_a(N,j,a):
    '''
    Params:
        N : int
            Number of spins in the system.
        j : int
            Spin index to place sigma_a, must be
            between 1 and N.
        a : str or int
            Wheter to use sigma_x, 'x' or 1, sigma_y,
            'y' or 2, or sigma_z, 'z' or 3.
    Output:
        Returns a Qobj containing the representation
        of the sigma_j_a operator in the base used for 
        defining the Pauli matrices. The definition of
        sigma_j_a is:
        
        sigma_j_a = I^(@j-1) @ sigma_a @ I^(@N-j)
    '''
    # Define I
    I = sigma_0
    # Select sigma_a
    if a == 'x' or a == 1:
        sigma_a = sigma_x
    elif a == 'y' or a == 2:
        sigma_a = sigma_y
    elif a == 'z' or a == 3:
        sigma_a = sigma_z
    
    # Build array
    operator = []
    for i in range(N):
        k = i+1
        if k <= j-1:
            operator.append(I)
        elif k == j:
            operator.append(sigma_a)
        else:
            operator.append(I)
            
    return Q.tensor(operator)


# 3 - DEFINE SYSTEM 
def generate_Heisenberg(N,J,h):
    '''
    Params:
        N : int
            Number of spins in the system.
        J : float or int
            Spin coupling constant.
        h : float or int
            External field coupling constant.
    Output:
        Returns a Qobj containing the Hamiltonian
        of the Heisenberg model, for N spins.
    '''
    # Define scale factor
    scale = -1/2
    
    # Initialize H
    H = sigma_j_a(N,1,'z') * 0
    
    # Build H
    for i in range(N):
        j = i+1
        row = sigma_j_a(N,1,'z') * 0
        # Add spin couplings with PBC
        for u in ['x','y','z']:
            if j == N:
                row += J*sigma_j_a(N,j,u)*sigma_j_a(N,1,u)
            else:
                row += J*sigma_j_a(N,j,u)*sigma_j_a(N,j+1,u)

        # Add external field coupling
        row += h*sigma_j_a(N,j,a='z')

        # Add to H
        H += row
        
    # Scale H and return
    return scale*H


# 3 - CALCULATIONS PIPELINE
def run_Heisenberg(N,J,h,verbose=0):
    '''
    Params:
        N : int
            Number of spins to include in the system.
        J : float
            Heisenberg model coupling constant between adyacent
            spins.
        h : float
            Heisenberg model external field coupling constant.
        verbose : int 0, 1 or 2
            Wheter to show current run status, 1, final run status, 2, 
            or no status, 0 (default).
    Output:
        Returns the following dictionary:
        'tensor'      -> Qobj object associated with the system 
        'eigenvalues' -> Eigenvalues of the system 
        'eigenstates' -> Eigenstates of the system
        'log'         -> Time log with the performance of the run.
    '''
    # Let's time every step and the overall performance
    t0 = time.time()
    # Build Hamiltonian as a CompositeOperator
    if verbose == 1:
        print('Building system...')
    
    H = generate_Heisenberg(N,J,h)
    
    t1 = time.time()
    if verbose == 1:
        print('Elapsed time: %.3f seconds' % (t1-t0))
        print('\nCalculating eigenvalues and eigenstates...')
    # Get eigenvalues
    eigen = H.eigenstates()
    eigenvalues = eigen[0]
    eigenstates = eigen[1]

    t2 = time.time()
    
    if verbose == 1:
        print('Elapsed time: %.3f seconds' % (t2-t1))
    if verbose == 1 or verbose == 2:
        print('\nProcess completed. Total wall time: %.3f seconds' % (t2-t0))
    
    # Build log to compare peformances
    time_log = {'Build':(t1-t0), 'Eigen':(t2-t1), 'Total':(t2-t0)}
    
    return {'tensor':H, 
            'eigenvalues':eigenvalues,
            'eigenstates':eigenstates,
            'log':time_log}

def plot_eigen(sim, energies=True, states=True, filepath=None):
    '''
    Params:
        sim : dict
            Dictionary containing at least the keys
            'eigenvalues' and 'eigenstates', with the
            information to be plotted.
        energies : bool (optional)
            Whether to plot the eigenvalues or not. Default
            is True.
        states : bool (optional)
            Whether to plot the eigenstates or not. Default
            is True.
        filepath : str or None (optional)
            Saves the Figures to the given path. Doesn't
            save when None (default).
    Output:
        Plots the eigenstates and/or eigenenergies of the 
        given system.
    '''
    if energies:
        # Init plot
        fig1, ax1 = plt.subplots(figsize=(5,4))
        ax1.hist(np.real(sim['eigenvalues']))
        plt.xticks(fontsize=15,rotation=20)
        plt.yticks(fontsize=15)
        plt.xlabel('Energy',fontsize=16)
        plt.ylabel('Count',fontsize=16)
        plt.tight_layout()
        if filepath != None:
            path = filepath.replace('.png', 
                                    '_energies.png')
            plt.savefig(path,dpi=200)
        plt.show();
        
    if states:
        # Get data
        data = np.real(np.hstack([
            np.array(state) for state in sim['eigenstates']
            ]))
        
        # Init plot
        FIGSIZE = 15
        LABELSIZE = 30
        fig2, ax2 = plt.subplots(figsize=(FIGSIZE,FIGSIZE))
        # Plot and color bar
        im = ax2.imshow(data, cmap='RdBu')
        cbar = ax2.figure.colorbar(im, ax=ax2, 
                                   fraction=0.046, pad=0.04)
        
        ax2.set_xticklabels([])
        ax2.set_yticklabels([])
        
        ax2.set_xlabel('Eigenstates',fontsize=LABELSIZE)
        ax2.set_ylabel('Base states',fontsize=LABELSIZE)
        
        cbar.ax.tick_params(labelsize=LABELSIZE)

        # Rotate the tick labels and set their alignment.
        plt.setp(ax2.get_xticklabels(), rotation=45, ha="right",
                 rotation_mode="anchor")

        plt.tight_layout()
        if filepath != None:
            path = filepath.replace('.png', 
                                    '_states.png')
            plt.savefig(path,dpi=200)
        plt.show();
        
    if states + energies == 0:
        print('Error: either states or energies must be True')
        
        
def compare_performance(logs_list, labels=None,
                        yscale='linear',xlabel=None, 
                        filepath=None):
    '''
    Params:
        logs_list : list
            List of run_Heisenberg()['log'] results.
        labels : list (optional)
            List of experiment labels.
        yscale : str (optional)
            Tipe of y-scale: 'linear', 'log', ...
        xlabel : str or None (optional)
            Label for x-axis, None by default.
        filepath : str or None (optional)
            Path to which save the plot, None by default.
    Output:
        Compares different the performance of different
        experiments.
    '''
    fig = plt.Figure(figsize=(20,15))
    # Plot all the data
    keys = ['Build','Eigen','Total']
    for i in range(3):
        data = []
        for log in logs_list:
            data.append(log[keys[i]])
        plt.plot(np.linspace(start=1,
                             stop=len(logs_list),
                             num=len(logs_list)),
                 data,'o--',label=keys[i])
    if labels != None:
        plt.xticks(np.linspace(start=1,
                             stop=len(logs_list),
                             num=len(logs_list)),
            labels=labels,fontsize=18)
    else:
        plt.xticks(np.linspace(start=1,
                             stop=len(logs_list),
                             num=len(logs_list)),
                   fontsize=18)
    plt.yscale(yscale)
    plt.yticks(fontsize=18)
    if xlabel != None:
        plt.xlabel(xlabel,fontsize=18)
    else:
        plt.xlabel('Experiment',fontsize=18)
    plt.ylabel(r'Time $[s]$',fontsize=18)
    plt.legend(fontsize=18)
    plt.tight_layout()
    if filepath != None:
        plt.savefig(filepath)
    plt.show();
```

# 7. Exploring different ways to characterize the state of the system

Let's define two important metrics that could help us describe the state of a system, both in macro and micro-scale. We define the pair correlation average value as:

$$C_{pair, z} = \frac{1}{N}\langle \sigma_i^z\sigma_{i+1}^z \rangle \ ,$$

and the magnetization over the $z$-axis as:

$$ M_z = \frac{1}{N} \langle \sigma_i^z \rangle$$

We'll explore a system with $N = 12$, $J = 2$ and $h=2$.

```python
N = 12
sim_explore = run_Heisenberg(N=N,J=2,h=2,verbose=2)
```

```python
plot_eigen(sim_explore,states=False)
```

```python
# Let's define the metrics we need
def C_pair_z(state, N):
    '''
    Params:
        state : Qobj
            State of the system.
        N : int
            Number of spins in the system.
    Output:
        Returns the average over the system state of 
        the pair correlation function over the z axis.
    '''
    # Get average value
    total = 0
    for i in range(N):
        j = i+1
        # Build operator
        if j == N:
            op = sigma_j_a(N,j,'z')*sigma_j_a(N,1,'z')
        else:
            op = sigma_j_a(N,j,'z')*sigma_j_a(N,j+1,'z')
        
        # Get inner prod and add to total
        total += Q.expect(op, state)
    return np.real(total/N)
    


def M_z(state, N):
    '''
    Params:
        state : Qobj
            State of the system.
        N : int
            Number of spins in the system.
    Output:
        Returns the average over the system state of 
        the magnetization over the z axis.
    '''
    # Get average value
    total = 0
    for i in range(N):
        j = i+1
        # Build operator
        op = sigma_j_a(N,j,'z')

        # Get inner prod and add to total
        total += Q.expect(op,state)
    return np.real(total/N)

    

def plot_two(col1,col2,kind='scatter',title=None,
             xlabel=None,ylabel=None,fontsize=16):
    '''
    Params:
        col1, col2 : list or numpy array
            Features x and y, respectively to plot.
        kind : str (optional)
            Type of plot. Default is scatter.
        title : str or None (optional)
            Title of the plot. Default is None.
        xlabel, ylabel : str or None (optional)
            Labels for the x and y axis, respectively.
            Default is None.
        fontsize : int (optional)
            Fonstize for every piece of text in the plot.
            Default is 16
    Output:
        Plots the features with the specified settings.
    '''
    # Allowed types of plot
    allowed = ['scatter']
    if kind == 'scatter':
        fig = plt.Figure(figsize=(15,10))
        plt.scatter(col1,col2)
        if title != None:
            plt.title(title,fontsize=fontsize)
        if ylabel != None:
            plt.ylabel(ylabel,fontsize=fontsize)
        if xlabel != None:
            plt.xlabel(xlabel,fontsize=fontsize)
        plt.xticks(rotation=25,fontsize=fontsize*0.9)
        plt.yticks(fontsize=fontsize*0.9)
        plt.show();
    else:
        print('Please select a valid type of plot: ' +\
              str(allowed))
```

```python
# Let's see this in action for the eigenstates 
# of the system
C_pair_list = []
M_list = []
for state in sim_explore['eigenstates']:
    C_pair_list.append(C_pair_z(state, N))
    M_list.append(M_z(state, N))
```

```python
plot_two(sim_explore['eigenvalues'],C_pair_list,
         xlabel=r'$E$',ylabel=r'$C_{pair}$')
plot_two(sim_explore['eigenvalues'],M_list,
        xlabel=r'$E$',ylabel=r'$M_z$')
plot_two(M_list,C_pair_list,
        xlabel=r'$M_z$',ylabel=r'$C_{pair}$')
```

Now that we've seen this works, let's see how the system behaves for different values of $h$.

```python
# Run multiple systems
sims_M = []
h_values_2 = [0.5,1,2,4,8]
for i in range(len(h_values_2)):
    h = h_values_2[i]
    print('\nRun %d of %d, h = %d' % (i+1,len(h_values_2),h))
    sim_i = run_Heisenberg(N=12,J=2,h=h,verbose=0)
    sims_M.append(sim_i)
```

```python
C_pair_all = []
M_z_all = []
for i in range(len(sims_M)):
    C_row_iter = []
    M_row_iter = []
    sim_i = sims_M[i]
    for state in sim_i['eigenstates']:
        C_row_iter.append(C_pair_z(state,N))
        M_row_iter.append(M_z(state,N))
    C_pair_all.append(C_row_iter)
    M_z_all.append(M_row_iter)
```

```python
fig,ax = plt.subplots(figsize=(30,75),nrows=5,ncols=2)
for i in range(len(C_pair_all)):
    title = r'$h=%s$'%str(h_values_2[i])
    ax[i][0].scatter(M_z_all[i],
                C_pair_all[i],
                s=50,
                label=title)
    ax[i][1].scatter(sims_M[i]['eigenvalues'],
                M_z_all[i],
                s=50,
                label=title)
    
    ax[i][0].legend(fontsize=30)
    ax[i][1].legend(fontsize=30)
    ax[i][0].set_ylabel(r'$C_{pair}$',fontsize=30)
    ax[i][1].set_ylabel(r'$M_z$',fontsize=30)
    ax[i][1].set_xlim(left=-60,right=40)
ax[-1][0].set_xlabel(r'$M_z$',fontsize=30)
ax[-1][1].set_xlabel(r'$E$',fontsize=30)
for axis in ax:
    for a in axis:
        a.tick_params(axis='x', rotation=45, 
                       labelsize=30)
        a.tick_params(axis='y', labelsize=30)
    
plt.tight_layout()
plt.show();
```
