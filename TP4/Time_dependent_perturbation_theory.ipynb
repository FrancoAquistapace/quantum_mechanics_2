```python
# Import necesary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import qutip as Q
```

# 1. Mathematical model

We are given the following Hamiltonian:

$$ \hat{H}(t) = \begin{bmatrix} 0 & \beta \\ \beta & 0 \end{bmatrix} + E_0 \sin(\omega t)\begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix} $$

And we have to analyze the system using various methods. We have to prepare the system in the fundamental state, this is the eigenstate $|\psi\rangle$ that satisfies the following:

$$\hat{H}(t=0)|\psi\rangle = \min\{\beta,-\beta\}|\psi\rangle$$

To make the notation easier we'll use $\hbar\equiv1$. 

We also define the dipole moment operator for this system as:

$$\mu \equiv -\frac{\partial H}{\partial (E_0\sin(\omega t))} = \sigma_z$$

## 1.1. Crank-Nicholson propagator:

In this method, we integrate the following equation of motion:

$$ |\psi(t+\Delta t)\rangle = \hat{U}(t)|\psi(t)\rangle \ , $$

using the Crank-Nicholson propagator:

$$ \hat{U}(t) = \frac{\hat{I} + (i/2\hbar)\hat{H}\Delta t}{\hat{I} - (i/2\hbar)\hat{H}\Delta t}$$


## 1.2. First order perturbation propagator:

In this method, we integrate the same equation of motion as before, using the first order perturbation propagator:

$$ \hat{U}(t) = \hat{I} + iE_0 \int_{0}^{t}e^{iH_0 \tau}\sin(\omega \tau)\sigma_z e^{-iH_0 \tau} d\tau = I + \frac{iE_0\sigma_z}{\omega}\bigl(1-\cos(\omega t)\bigr) + 2iE_0\sigma_y\bigl(\Lambda_1(t)+\Lambda_2(t)\sigma_x\bigr)$$

where

$$\Lambda_1(t) = \frac{1}{4}\biggl[\frac{\sin[t(\omega-2\beta)]}{\omega-2\beta} - \frac{\sin[t(\omega+2\beta)]}{\omega+2\beta}\biggr] $$

and

$$\Lambda_2(t) = \frac{i}{4}\biggl[\frac{\bigl[1-\cos[t(\omega+2\beta)\bigr]}{\omega+2\beta}+\frac{\bigl[1-\cos[t(\omega-2\beta)\bigr]}{\omega-2\beta}+\frac{2}{\omega}\bigl[\cos(\omega t)-1\bigr]\biggr]$$

In this method we have the following response function for the dipole moment $\mu$:

$$\phi_{\mu}(t) =2\sin(2|\beta|t)$$

so that the value $\langle\mu\rangle(t)$ is given by:

$$\langle\mu\rangle(t) = \frac{2E_0}{(\omega^2 -4\beta^2)}\bigl[\omega\sin(2|\beta|t)-2|\beta|\sin(\omega t)\bigr] $$

Note that this function is undefined at $\omega=2|\beta|$, but can be extended by taking the limit $\omega\rightarrow2|\beta|$:

$$\left.\langle\mu\rangle(t)\right|_{\omega=2|\beta|} = \frac{E_0}{2|\beta|}\bigl[\sin(2|\beta|t)-2|\beta|t\cos(2|\beta|t)\bigr]$$

On the other hand, the response function for $H(t)$ is:

$$\phi_{H\mu}(t) = -2E_0\sin(\omega t)\sin(2|\beta|t) $$

so that the value $\langle H\rangle(t)$ is given by:

$$\langle H\rangle(t) = -|\beta| - \frac{E_0^2\omega}{\omega^2-\beta^2}\biggl[\frac{\omega}{|\beta|}\cos(\omega t)\bigl[\cos(2|\beta|t)-1\bigr]+\sin(\omega t)\sin(2|\beta|t)\biggr]$$


## 1.3 Fermi Golden Rule

For the given system, preparing it in the fundamental state, the state evolves as:

$$ |\psi(t)\rangle = |\psi_0\rangle + c(t)|\psi_1\rangle$$

where $|\psi_0\rangle$ is the fundamental state and $|\psi_1\rangle$ is the excited state. The coefficient $c$ is given by:

$$c(t) = \frac{-E_0}{2i}\biggl[\frac{e^{it(\omega+2|\beta|)}-1}{\omega+2|\beta|}+\frac{e^{-it(\omega-2|\beta|)}-1}{\omega-2|\beta|}\biggr]$$

Note that we have

$$ P_{fundamental}(t) = 1$$

and

$$ P_{excited}(t) = |c(t)|^2 = \biggl(\frac{E_0\omega}{\omega^2 - 4\beta^2}\biggr)^2\biggl[\cos^2(\omega t)+\frac{4\beta^2}{\omega^2}\sin^2(\omega t)-2\biggl(\cos(\omega t)\cos(2|\beta|t)+\frac{2|\beta|}{\omega}\sin(\omega t)\sin(2|\beta|t)\biggr)\biggr] $$

For the expectation value $\langle\mu\rangle(t)$, we have:

$$\langle\mu\rangle(t) = 2\Re[c(t)]=\frac{2E_0\omega}{\omega^2-4\beta^2}\biggr(\frac{2|\beta|}{\omega}\sin(\omega t)\sin(2|\beta|t) - \cos(\omega t)\cos(2|\beta|t)\biggl)$$

Finally, we have:

$$\langle H\rangle(t) = -|\beta| + P_{excited}(t)|\beta|-E_0\sin(\omega t)\langle\mu\rangle(t)$$


## 1.4 Rotating Wave Approximation (RWA)

We take the same path as in the Fermi golden rule, but taking the following approximation:

$$ |\omega-2|\beta||\ll |2\beta| $$

Now, we have

$$c(t) = \frac{-E_0}{2i}\biggl(\frac{e^{-it(\omega-2|\beta|)}-1}{\omega-2|\beta|}\biggr)$$

As in the FGR, we have

$$ P_{fundamental}(t) = 1$$

and

$$ P_{excited}(t) = |c(t)|^2 = \biggl|\frac{E_0\sin((\omega-2|\beta|)t)}{\omega-2|\beta|}\biggr|^2$$

For the expectation value $\langle\mu\rangle(t)$, we have:

$$\langle\mu\rangle(t) = 2\Re[c(t)]= \frac{E_0\sin((\omega-2|\beta|)t)}{\omega-2|\beta|}$$

Finally, we have the same expression of $\langle H\rangle(t)$ as a function of $P_{excited}$ and $\langle\mu\rangle(t)$ as in the FGR, although it differs when considered as a function of $t$:

$$\langle H\rangle(t) = -|\beta| + P_{excited}(t)|\beta|-E_0\sin(\omega t)\langle\mu\rangle(t)$$

# 2. Finding the fundamental state


We know that the Hamiltonian at time $t=0$ is given by:

$$ \hat{H}(t=0) = \begin{bmatrix} 0 & \beta \\ \beta & 0 \end{bmatrix}$$

This is diagonalized by the following eigenstates:

$$ |\beta\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix} \ \ \text{and} \ \ 
   |-\beta\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix} \ \ ,$$
   
with eigenvalues $\beta$ and $-\beta$ respectively. So the fundamental and excited states are respectively, $|\min\{\beta,-\beta\}\rangle$ and $|\max\{\beta,-\beta\}\rangle$.

```python
# Let's define a function that can generate the fundamental 
# state of the system for any value of beta
def get_fundamental(beta):
    c = 1/np.sqrt(2)
    if beta <= 0:
        return Q.Qobj(c*np.array([[1],[1]]))
    elif beta > 0:
        return Q.Qobj(c*np.array([[1],[-1]]))
    

# We should also define a function that generates
# the excited state
def get_excited(beta):
    c = 1/np.sqrt(2)
    if beta <= 0:
        return Q.Qobj(c*np.array([[1],[-1]]))
    elif beta > 0:
        return Q.Qobj(c*np.array([[1],[1]]))
    
    
# Define the dipole moment operator
mu = Q.sigmaz()
```

# 3. Designing a pipeline for the simulations

```python
# CRANK-NICHOLSON PIPELINE
# Define a function that returns H
def H(beta, E_0, omega, t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter for H.
    Output:
        Returns the Hamiltonian of the system.
    '''
    return beta*Q.sigmax()-E_0*np.sin(omega*t)*Q.sigmaz()

# Define Crank-Nicholson propagator
def CN_U(delta_t, H):
    '''
    Params:
        delta_t : float
            Time interval to use for the propagator.
        H : QuTip Qobj
            H operator of the system.
    Output:
        Returns the Crank-Nicholson propagator of the system.
    '''
    top = Q.qeye(2)+(1j/2)*H*delta_t
    bottom = Q.qeye(2)-(1j/2)*H*delta_t
    return top*bottom.inv()


# Define Crank-Nicholson Pipeline
class CN_Pipeline():
    def __init__(self,delta_t,beta, E_0, omega):
        '''
        Params:
            delta_t : float
                Time interval to be applied in between each step of the integration.
            beta : float
                Parameter for the time independent term of H.
            E_0 : float
                Maximum energy for the time dependent term of H.
            omega : float
                Angular frequency for the time dependent term of H.
                
        Output:
            Initializes the Pipeline object.
        '''
        self.delta_t = delta_t
        self.beta = beta
        self.E_0 = E_0
        self.omega = omega
        
        # Start time parameter
        self.t = 0
        
        # We get the fundamental state of the system
        self.state = get_fundamental(beta)
        # Now we build the Hamiltonian 
        self.H = H(beta, E_0, omega, t=self.t)
        # And the propagator
        self.U = CN_U(delta_t, self.H)
        
        # Initialize history of the system 
        self.history = [get_fundamental(beta)]
        
        # Initialize history of the expected value for H
        # and mu
        self.H_history = [Q.expect(self.H,self.state)]
        self.mu_history = [Q.expect(mu, self.state)]
        

    def step(self):
        '''
        Output:
            Updates the state of the system using the Crank-Nicholson
            approximation of the Propagator, saves new state to the
            system history and updates the value of t.
        '''
        # Update time
        self.t += self.delta_t
        # Redefine H and U
        self.H = H(self.beta, self.E_0, 
                   self.omega, t=self.t)
        self.U = CN_U(self.delta_t, self.H)
        # Get new state
        old_state = self.state
        new_state = self.U*old_state
        self.state = new_state.copy()
        # Add new state to system history
        self.history.append(new_state)
        # Add new expected energy value
        self.H_history.append(Q.expect(self.H,self.state))
        self.mu_history.append(Q.expect(mu, self.state))
        
        
    def run(self,steps):
        '''
        Params:
            steps : int
                Number of timesteps to run the pipeline.
        Output:
            Evolves the system a time t given by:
                t = steps*self.delta_t
        '''
        for i in range(steps):
            self.step()
        
        
# -------------------

# FIRST ORDER PERTURBATION PIPELINE

# Some auxiliary functions: Lambda 1 and 2
def Lambda_1(beta, omega, t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of Lambda_1(t) aas defined in the 
        first order perturbation framework.
    '''
    res = omega-2*beta
    anti = omega+2*beta
    if np.isclose(res,0):
        return (1/4)*(t-(np.sin(t*anti)/anti))
    elif np.isclose(anti,0):
        return (-1/4)*t + (1/4)*(np.sin(t*res)/res)
    else:
        res_term = np.sin(t*(res))/res
        anti_res_term = np.sin(t*anti)/anti
        return (1/4)*(res_term - anti_res_term)

def Lambda_2(beta, omega, t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of Lambda_2(t) aas defined in the 
        first order perturbation framework.
    '''
    res = omega-2*beta
    anti = omega+2*beta
    if np.isclose(res,0):
        anti_res_term = (1 - np.cos(t*anti))/anti
        omega_term = 2*(np.cos(omega*t)-1)/omega
        return (1j/4)*(anti_res_term + omega_term)
    elif np.isclose(anti,0):
        res_term = (1 - np.cos(t*res))/res
        omega_term = 2*(np.cos(omega*t)-1)/omega
        return (1j/4)*(res_term + omega_term)
    else:
        res_term = (1 - np.cos(t*res))/res
        anti_res_term = (1 - np.cos(t*anti))/anti
        omega_term = 2*(np.cos(omega*t)-1)/omega
        return (1j/4)*(res_term + anti_res_term + omega_term)


def FOP_U(beta, E_0, omega, t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.
    Output:
        Returns the first order perturbation propagator 
        of the system.
    '''
    term_1 = Q.qeye(2)
    term_2 = 1j*E_0*Q.sigmaz()*(1-np.cos(omega*t))/omega
    term_3 = 2j*E_0*Q.sigmay()*(Lambda_1(beta, omega, t)+Lambda_2(beta, omega, t)*Q.sigmax())
    return term_1 + term_2 + term_3


# Define First Order Perturbation Pipeline
class FOP_Pipeline():
    def __init__(self,delta_t,beta, E_0, omega):
        '''
        Params:
            delta_t : float
                Time interval to be applied in between each step of the integration.
            beta : float
                Parameter for the time independent term of H.
            E_0 : float
                Maximum energy for the time dependent term of H.
            omega : float
                Angular frequency for the time dependent term of H.
                
        Output:
            Initializes the Pipeline object.
        '''
        self.delta_t = delta_t
        self.beta = beta
        self.E_0 = E_0
        self.omega = omega
        
        # Start time parameter
        self.t = 0
        
        # We get the fundamental state of the system
        self.initial = get_fundamental(beta)
        self.state = get_fundamental(beta)
        # Now we build the Hamiltonian 
        self.H = H(beta, E_0, omega, t=self.t)
        # And the propagator
        self.U = FOP_U(beta, E_0, omega, self.t)
        
        # Initialize history of the system 
        self.history = [get_fundamental(beta)]
        
        # Initialize history of the expected value for H
        # and mu
        self.H_history = [Q.expect(self.H,self.state)]
        self.mu_history = [Q.expect(mu, self.state)]
        

    def step(self):
        '''
        Output:
            Updates the state of the system using the first order
            perturbation approximation of the Propagator, saves new
            state to the system history and updates the value of t.
        '''
        # Update time
        self.t += self.delta_t
        # Redefine H and U
        self.H = H(self.beta, self.E_0, 
                   self.omega, t=self.t)
        self.U = FOP_U(self.beta, self.E_0, self.omega, self.t)
        # Get new state
        old_state = self.state
        new_state = self.U*self.initial
        self.state = new_state.copy()
        # Add new state to system history
        self.history.append(new_state)
        # Add new expected energy value
        self.H_history.append(Q.expect(self.H,self.state))
        self.mu_history.append(Q.expect(mu, self.state))
        
        
    def run(self,steps):
        '''
        Params:
            steps : int
                Number of timesteps to run the pipeline.
        Output:
            Evolves the system a time t given by:
                t = steps*self.delta_t
        '''
        for i in range(steps):
            self.step()
        

# Let's define a function that calculates <mu>(t)
def avg_mu(beta, E_0, omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of <mu>(t) according to the linear
        response function in the first order perturbation
        framework.
    '''
    if np.isclose(omega-2*np.abs(beta),0):
        C = E_0/(2*np.abs(beta))
        return C*(np.sin(omega*t)-omega*t*np.cos(omega*t))
    else:
        C = 2*E_0/((omega**2) - 4*(np.abs(beta)**2))
        return C*(omega*np.sin(2*np.abs(beta)*t) - 2*np.abs(beta)*np.sin(omega*t))
    
    
# Function that calculates <H>(t)
def avg_H(beta, E_0, omega, t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of <H>(t) according to the linear
        response function in the first order perturbation
        framework.
    '''
    if np.isclose(omega-np.abs(beta),0):
        return 0*t
    else:
        C = (E_0**2)*omega/((omega**2) - (np.abs(beta)**2))
        aux_f = (omega/np.abs(beta))*np.cos(omega*t)*(np.cos(2*np.abs(beta)*t)-1)+\
                np.sin(omega*t)*np.sin(2*np.abs(beta)*t)
        return -np.abs(beta) - C*aux_f

    
    
# -------------------

# Fermi Golden Rule Pipeline
# To avoid any issues, we'll define c(t)
# also at the resonance case
def c_t(beta,E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of c(t) according to the Fermi
        golden rule (FGR) framework.
    '''
    if np.isclose(omega-2*np.abs(beta),0):
        C = -E_0/(2j)
        return C*((1/4*np.abs(beta))*(np.cos(4*np.abs(beta)*t) - 1 + 1j*np.sin(4*np.abs(beta)*t)) - 1j*t)
    else:
        omega_f = 2*np.abs(beta)
        C = -E_0*omega/(1j*(omega**2 - 4*(beta**2)))
        exp = np.cos(omega_f*t) + 1j*np.sin(omega_f*t)
        return C*((np.cos(omega*t)-1j*(omega_f/omega)*np.sin(omega*t))*exp - 1)
    

    
# Define observables 
def FGR_excited(beta,E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of P_excited according to the Fermi
        golden rule (FGR) framework.
    '''
    return np.power(np.abs(c_t(beta,E_0,omega,t)),2)

def FGR_mu(beta,E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of <mu>(t) according to the Fermi
        golden rule (FGR) framework.
    '''
    return 2*np.real(c_t(beta,E_0,omega,t))


def FGR_H(beta,E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of <H>(t) according to the Fermi
        golden rule (FGR) framework.
    '''
    return np.abs(beta)*(FGR_excited(beta,E_0,omega,t)-1) - E_0*np.sin(omega*t)*FGR_mu(beta,E_0,omega,t)


# -------------------

# RWA Pipeline
# To avoid any issues, we'll define c(t)
# also at the resonance case
def RWA_c_t(beta, E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of c(t) according to the Fermi
        golden rule (FGR) framework with RWA.
    '''
    delta_w = omega - 2*np.abs(beta)
    if np.isclose(delta_w,0):
        return -(E_0/2j)*(-1j*t)
    else:
        exp = np.cos(delta_w*t)-1j*np.sin(delta_w*t)
        return -(E_0/2j)*(exp - 1)/delta_w
    
    
# Define observables 
def RWA_excited(beta,E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of P_excited according to the Fermi
        golden rule (FGR) framework with RWA.
    '''
    return np.power(np.abs(RWA_c_t(beta,E_0,omega,t)),2)

def RWA_mu(beta,E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of <mu>(t) according to the Fermi
        golden rule (FGR) framework with RWA.
    '''
    return 2*np.real(RWA_c_t(beta,E_0,omega,t))


def RWA_H(beta,E_0,omega,t):
    '''
    Params:
        beta : float
            Parameter for the time independent term of H.
        E_0 : float
            Maximum energy for the time dependent term of H.
        omega : float
            Angular frequency for the time dependent term of H.
        t : float
            Time parameter.

    Output:
        Returns the value of <H>(t) according to the Fermi
        golden rule (FGR) framework with RWA.
    '''
    return np.abs(beta)*(RWA_excited(beta,E_0,omega,t)-1) - E_0*np.sin(omega*t)*RWA_mu(beta,E_0,omega,t)
    
    

# -------------------

# AUXILIARY FUNCTIONS
# Define auxiliary functions that can help us visualize the results
def get_probs(pipeline,mode='fundamental'):
    '''
    Params:
        pipeline : Pipeline
            Pipeline containing the simulation results in history attribute.
        mode : str
            Projection mode. 'fundamental' gets the fundamental state 
            probability, while 'excited' gets the excited state probability.
    
    Output:
        Returns a numpy array of the requested probabilities, for every step of 
        the pipeline history.
    '''
    # Let's get the proyection state
    if mode == 'fundamental':
        state = get_fundamental(pipeline.beta)
    elif mode == 'excited':
        state = get_excited(pipeline.beta)
    else:
        print('Error: mode must be either "fundamental" or "excited".')
        return
    
    # Get probabilities
    probs_list = []
    for psi in pipeline.history:
        prob = np.absolute(np.array(state.dag()*psi)[0][0])
        probs_list.append(prob)
        
    # Turn into numpy array and get squares
    probs_list = np.power(np.array(probs_list),2)
    return probs_list

    
    
def compare_histories(pipelines, mode='all', labels=None,
                      filepath=None):
    '''
    Params:
        pipelines : list
            List of Pipeline objects, containing histories to be
            compared.
        mode : str
            Plotting mode. 'fundamental' plots the fundamental state probability,
            'excited' plots the excited state probability, 'module' plots the module 
            of the state vector, 'energy' plots the energy of the system, 'mu' plots 
            the expected values of the dipole moment, and 'all' plots all the 
            previous options.
        labels : list or None
            List of str labels, with same amount of elements as pipelines. If
            None, then predetermined labels are used.
        filepath : str or None
            File path to save the generated plots, if None then the plots aren't saved.
        
    Output:
        Plots the requested information, comparing the provided pipelines.
    '''
    if labels == None:
        labels = []
        i = 1
        for p in pipelines:
            message = 'Pipeline ' + str(i)
            labels.append(message)
            i+=1
            
    if mode == 'fundamental' or mode == 'all':
        fig1 = plt.figure(figsize=(13,8))
        i = 0
        for p in pipelines:
            # Get probs for fundamental state
            probs = get_probs(p,mode='fundamental')
            plt.plot((p.delta_t)*np.linspace(0,len(probs),num=len(probs)), 
                     probs, label=labels[i], linewidth=2)
            i += 1
            if i == len(pipelines):
                dom = (p.delta_t)*np.linspace(0,len(probs),num=len(probs))
                plt.plot(dom, np.ones(shape=dom.shape),'--',
                         label=r'Linear response',linewidth=2)
                plt.plot(dom, np.ones(shape=dom.shape),'-.',
                         label=r'FGR',linewidth=2)
                plt.plot(dom, np.ones(shape=dom.shape),':',
                         label=r'FGR - RWA',linewidth=2)
        plt.xlabel('t',fontsize=20)
        plt.ylabel('$P_{fundamental}$', fontsize=20)
        plt.xticks(size=18)
        plt.yticks(size=18)
        plt.xlim(left=0)
        #plt.ylim(bottom=0)
        plt.legend(fontsize=20)
        
        if filepath != None:
            plt.savefig('fundamental_'+filepath)
        
    if mode == 'excited' or mode == 'all':
        fig2 = plt.figure(figsize=(13,8))
        i = 0
        for p in pipelines:
            # Get probs for fundamental state
            probs = get_probs(p,mode='excited')
            plt.plot((p.delta_t)*np.linspace(0,len(probs),num=len(probs)), 
                     probs, label=labels[i], linewidth=2)
            i += 1
            if i == len(pipelines):
                dom = (p.delta_t)*np.linspace(0,len(probs),num=len(probs))
                plt.plot(dom, np.zeros(shape=dom.shape),'--',
                         label=r'Linear response',linewidth=2)
                plt.plot(dom, FGR_excited(p.beta,p.E_0,p.omega,dom),'-.',
                         label=r'FGR',linewidth=2)
                plt.plot(dom, RWA_excited(p.beta,p.E_0,p.omega,dom),':',
                         label=r'FGR - RWA',linewidth=2)
        plt.xlabel('t',fontsize=20)
        plt.ylabel('$P_{excited}$', fontsize=20)
        plt.xticks(size=18)
        plt.yticks(size=18)
        plt.xlim(left=0)
        plt.ylim(bottom=0)
        plt.legend(fontsize=20)
        
        if filepath != None:
            plt.savefig('excited_'+filepath)
            
    if mode == 'module' or mode == 'all':
        fig5 = plt.figure(figsize=(13,8))
        i = 0
        for p in pipelines:
            # Get probs for fundamental state
            probs = get_probs(p,mode='excited')+get_probs(p,mode='fundamental')
            plt.plot((p.delta_t)*np.linspace(0,len(probs),num=len(probs)), 
                     probs, label=labels[i], linewidth=2)
            i += 1
            if i == len(pipelines):
                dom = (p.delta_t)*np.linspace(0,len(probs),num=len(probs))
                plt.plot(dom, np.ones(shape=dom.shape),'--',
                         label=r'Linear response',linewidth=2)
                plt.plot(dom, 1+FGR_excited(p.beta,p.E_0,p.omega,dom),'-.',
                         label=r'FGR',linewidth=2)
                plt.plot(dom, 1+RWA_excited(p.beta,p.E_0,p.omega,dom),':',
                         label=r'FGR - RWA',linewidth=2)
        plt.xlabel('t',fontsize=20)
        plt.ylabel('$P_{total}$', fontsize=20)
        plt.xticks(size=18)
        plt.yticks(size=18)
        plt.xlim(left=0)
        plt.legend(fontsize=20)
        
        if filepath != None:
            plt.savefig('mod_'+filepath)
        
    if mode == 'energy' or mode == 'all':
        fig3 = plt.figure(figsize=(13,8))
        i = 0
        for p in pipelines:
            # Get energies
            energies = p.H_history
            plt.plot((p.delta_t)*np.linspace(0,len(energies),num=len(energies)), 
                     energies, label=labels[i], linewidth=2)
            i += 1
            if i == len(pipelines):
                dom = (p.delta_t)*np.linspace(0,len(energies),num=len(energies))
                plt.plot(dom, avg_H(p.beta,p.E_0,p.omega,dom),'--',
                         label=r'Linear response',linewidth=2)
            
                plt.plot(dom, FGR_H(p.beta,p.E_0,p.omega,dom),'-.',
                         label=r'FGR',linewidth=2)
                plt.plot(dom, RWA_H(p.beta,p.E_0,p.omega,dom),':',
                         label=r'FGR - RWA',linewidth=2)
                
        plt.xlabel('t',fontsize=20)
        plt.ylabel(r'$\langle H\rangle$', fontsize=20)
        plt.xticks(size=18)
        plt.yticks(size=18)
        plt.xlim(left=0)
        #plt.ylim(bottom=0)
        plt.legend(fontsize=20)
        
        if filepath != None:
            plt.savefig('H_'+filepath)
            
    if mode == 'mu' or mode == 'all':
        fig4 = plt.figure(figsize=(13,8))
        i = 0
        for p in pipelines:
            # Get mu values
            mu_vals = p.mu_history
            plt.plot((p.delta_t)*np.linspace(0,len(mu_vals),num=len(mu_vals)), 
                     mu_vals, label=labels[i], linewidth=2)
            i += 1
            if i == len(pipelines):
                dom = (p.delta_t)*np.linspace(0,len(mu_vals),num=len(mu_vals))
                plt.plot(dom, avg_mu(p.beta, p.E_0, p.omega,dom),'--',
                         label=r'Linear response',linewidth=2)
                
                plt.plot(dom, FGR_mu(p.beta,p.E_0,p.omega,dom),'-.',
                         label=r'FGR',linewidth=2)
                plt.plot(dom, RWA_mu(p.beta,p.E_0,p.omega,dom),':',
                         label=r'FGR - RWA',linewidth=2)
                
        plt.xlabel('t',fontsize=20)
        plt.ylabel(r'$\langle \mu\rangle$', fontsize=20)
        plt.xticks(size=18)
        plt.yticks(size=18)
        plt.xlim(left=0)
        #plt.ylim(bottom=0)
        plt.legend(fontsize=20)
        
        if filepath != None:
            plt.savefig('mu_'+filepath)
            
        
    plt.show()
    
    
# Finally, a function to run everything
def run_all(beta,E_0,omega,delta_t_list=[0.001],t_max=20):
    # Let's prepare some delta_t values, we'll use some fixed constants
    # for the rest of the parameters
    BETA, E_0, OMEGA = beta, E_0, omega

    CN_sims = [CN_Pipeline(delta_t=delta_t, 
                           beta=BETA,E_0=E_0, 
                           omega=OMEGA) for delta_t in delta_t_list]

    # Let's run the simulations, adjusting the steps so that the 
    # time duration is the same for every sim
    for i in range(len(delta_t_list)):
        steps = int(t_max/delta_t_list[i])
        cn_sim = CN_sims[i]
        cn_sim.run(steps)
        
    # Now let's run the FOP method
    FOP_sims = [FOP_Pipeline(delta_t=delta_t, 
                           beta=BETA,E_0=E_0, 
                           omega=OMEGA) for delta_t in delta_t_list]

    # Let's run the simulations, adjusting the steps so that the 
    # time duration is the same for every sim
    for i in range(len(delta_t_list)):
        steps = int(t_max/delta_t_list[i])
        fop_sim = FOP_sims[i]
        fop_sim.run(steps)
        
    # Build labels and plot
    all_sims = [*CN_sims,*FOP_sims]
    all_labels = [r'CN - $\Delta t = %.3f$' %(p.delta_t) for p in CN_sims]
    all_labels.extend([r'FOP - $\Delta t = %.3f$' % (p.delta_t) for p in FOP_sims])

    compare_histories(all_sims, labels=all_labels)
```
